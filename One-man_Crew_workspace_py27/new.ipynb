## Graph creation

```python
def graph_create(input, terminal_loc = [0], shift_hr = 24):
    
    G = nx.DiGraph() #creates graph
    
    G.add_nodes_from(input) #adds nodes from xy
    terminal_begin = set() # disordered list
    terminal_end = set()
    for ind, t in enumerate(terminal_loc): # ind: index; t: value
        G.add_nodes_from([('T'+str(ind)+'Begin', {'coord':(t,0),'property' : 'T','label' : 'T'+str(ind)+'Begin'})]) # nx.get_node_attributes(G,'label')
        G.add_nodes_from([('T'+str(ind)+'End', {'coord' : (t,shift_hr), 'property' : 'T', 'label' : 'T'+str(ind)+'End'})])
        
        terminal_begin.add('T'+str(ind)+'Begin')
        terminal_end.add('T'+str(ind)+'End')
    for (n,m) in it.permutations(G.nodes(),2):
        if n in terminal_end:
            continue
        else:
            if G.nodes[n]['coord'][1] < G.nodes[m]['coord'][1]: # ['coord'][1] 是时间，后一个点要比前一个点晚，才能被visit
                if not ((n in terminal_begin) == (m in terminal_end) == True): # 不能两个点都在terminal（terminal之间没有连线）
    
                    G.add_edge(n, m,
                               dist     = abs(G.nodes[n]['coord'][0] - G.nodes[m]['coord'][0]), 
                               delt_t   = abs(G.nodes[n]['coord'][1] - G.nodes[m]['coord'][1]),
                               spd      = round(abs(G.nodes[n]['coord'][0] - G.nodes[m]['coord'][0]) / abs(G.nodes[n]['coord'][1] - G.nodes[m]['coord'][1]),2)
                               #speed in miles per hour
                               )
                    
    
    return G
```

```python
def transfer_graph_to_solution_dic(G):
    G_c = G.copy() # G_c 的改变不会影响原始的G
    terminal_begin_list = []
    terminal_end_list = []
    for n in G_c.nodes():
        if G_c.nodes[n]['property'] == 'T': # 选出是terminal的
            if n[2] == 'B':
                terminal_begin_list.append(n)
            if n[2] == 'E':
                terminal_end_list.append(n)

    terminal_begin_list.sort()
    terminal_end_list.sort()
    solution_dic = {}

    for i in range(len(terminal_begin_list)):
        solution_dic[terminal_begin_list[i]] = []
        for path in nx.all_simple_paths(G_c, source = terminal_begin_list[i], target=terminal_end_list[i]): # 选出所有可能的path
            solution_dic[terminal_begin_list[i]].append(path)
    
    return solution_dic
```

## Construction heuristic

```python
def constr_heu_srch(n_ind, G, dist_first = True):
    term_end_list = []
    term_begin_list = []
    for n in G.nodes():
        if G.nodes[n]['property'] == 'T':
            if G.out_degree(n) == 0: # 这个点接下来没有连着其他点
                term_end_list.append(n)
            if G.in_degree(n) == 0:
                term_begin_list.append(n) # 这个点前面没有其他点
                
    candi_list = []
    candi_dist_list = []
    candi_delt_list = [] 
    next_node = [None, None]
    if G.successors(n_ind): # Returns an iterator over successor nodes of n.
                            #A successor of n is a node m such that there exists a directed edge from n to m.
        for nb in G.successors(n_ind): 
            if nb in term_end_list: # 如果这个node是end，就不要它
                continue
            else:
                candi_list.append(nb) #如果这个node不是end，就成为candidate
                candi_dist_list.append(G[n_ind][nb]['dist']) # not G.edges[node1][node2]['dist'] but G[node1][node2]['dist']
                candi_delt_list.append(G[n_ind][nb]['delt_t'])
        if dist_first == True:
            if candi_dist_list:
                next_node[0] = candi_list[candi_dist_list.index(min(candi_dist_list))]
                next_node[1] = G[n_ind][nb]['dist'] / G[n_ind][nb]['delt_t'] 
        if dist_first == False:
            if candi_delt_list:
                next_node[0] = candi_list[candi_delt_list.index(min(candi_delt_list))]
                next_node[1] = G[n_ind][nb]['dist'] / G[n_ind][nb]['delt_t'] 
    return next_node
```

```python
def construct_heu_feasi_sol(G, max_speed = None, dist_first = True, rand = False):   
    print ('    ****Construction Heuristic')
    solution = {}
    traversed_nodes = set()
    out_only_nodes =set()
    in_only_nodes = set()
    island_nodes = set()
    terminal_nodes = set()
    term_begin_list = []
    term_end_list = []
    shift_hr = 0
    for n in G.nodes():
        if G.nodes[n]['property'] == 'T':
            if n[2] == 'E':
                term_end_list.append(n)
                shift_hr = G.nodes[n]['coord'][1] # update shifthour
            if n[2] == 'B':
                term_begin_list.append(n)  
    term_end_list.sort()
    term_begin_list.sort()
    for ind in range(len(term_begin_list)):
        solution[term_begin_list[ind]] = []
        
    G_c = G.copy() # G_c 的改变不会影响原始的G
    
    if max_speed:
        for (e1,e2) in G.edges(): #如果用G_c.edges()则会出现dictionary changed size during iteration的error
            if G_c[e1][e2]['spd'] > max_speed: #是G_c[e1][e2]['spd']不是G_c.edge[e1][e2]['spd']
                G_c.remove_edge(e1,e2)
    
    for n in G.nodes():
        if G_c.nodes[n]['property'] == 'WE':
            if G_c.nodes[n]['coord'][1] > shift_hr:
                G_c.remove_node(n)

    
    for n in G_c.nodes():
        if G_c.in_degree(n) == 0 and G_c.in_degree(n) != G_c.out_degree(n):
            out_only_nodes.add(n)
        if G_c.out_degree(n) == 0 and G_c.in_degree(n) != G_c.out_degree(n):
            in_only_nodes.add(n)
        if G_c.in_degree(n) == G_c.out_degree(n) == 0:
            island_nodes.add(n)
    
    
    if rand:        
        a = range(len(term_begin_list))
#         random.shuffle(a) #不知道他想shuffle啥
        term_begin_list = [term_begin_list[i] for i in a]
        term_end_list = [term_end_list[i] for i in a]
    
    #-----------------------------------------------#
    indicator = 0
    while set(G_c.nodes())-set(term_begin_list)-set(term_end_list)!= traversed_nodes and indicator < len(term_begin_list):
        if rand:        
            a = range(len(term_begin_list))
#             random.shuffle(a)
            term_begin_list = [term_begin_list[i] for i in a]
            term_end_list = [term_end_list[i] for i in a]
        indicator = 0 
        for ind in range(len(term_begin_list)):
            ind = random.choice(range(len(term_begin_list)))
            if not nx.has_path(G_c, source=term_begin_list[ind], target=term_end_list[ind]):
                indicator += 1
    
            for tn in traversed_nodes:
                if tn in G_c.nodes():
                    G_c.remove_node(tn)
            path = [term_begin_list[ind]]
            flag_t = path[-1][:2]
            
            while len(list(G.successors(path[-1])))>0: #list of neighbors # 否则会有object of type 'dict_keyiterator' has no len()
                flag = path[-1]
                next_heu_candi = constr_heu_srch(path[-1], G = G_c, dist_first=dist_first)
                if (next_heu_candi[0], term_end_list[ind]) in G_c.edges():
                    if next_heu_candi[0] == term_end_list[ind]:
                        path.append(next_heu_candi[0])
                    if (path[-1], next_heu_candi[0]) in G_c.edges():
                        path.append(next_heu_candi[0])
                else:
                    if (path[-1], term_end_list[ind]) in G_c.edges():
                        path.append(term_end_list[ind])
                if not next_heu_candi[0]:
                    if (path[-1],term_end_list[ind]) in G_c.edges():
                        if G[path[-1]][term_end_list[ind]]['spd'] <= max_speed:
                            path.append(term_end_list[ind])
                
                if type(path[-1]) is str:
                    if path[-1][:2] == flag_t:
                        break
                if path[-1] == flag:
                    break
            
            if path[-1] != term_end_list[ind]:
                flag_2 = 0
                while flag_2 == 0 and len(path)>1:
                    path.pop(-1)
                    if (path[-1], term_end_list[ind]) in  G_c.edges():
                        path.append(term_end_list[ind])
                        flag_2 = 1

            


            if path[-1] == path[0]:
                if constr_heu_srch(path[-1], G_c, dist_first=dist_first)[0]:
                    G_c.remove_edge(path[-1],constr_heu_srch(path[-1], G_c, dist_first=dist_first)[0])
            if path[0] in term_begin_list and path[-1] in term_end_list and len(path)>2 and path[0][1] == path[-1][1]:
                for t_n in path[1:-1]:
                    if t_n not in terminal_nodes:
                        traversed_nodes.add(t_n)            

                solution[term_begin_list[ind]].append(path)
                    
        
        #-----------------------------------------------#

            
    visited = set()
    t_c_number = {}
    
    for t,p in solution.items():
        t_c_number[t] = len(p)
        for n in p:
            for i in n:
                visited.add(i)

    t_nmbr = 0
    for t in t_c_number:
        t_nmbr += t_c_number[t]
    
    return_list = [solution, 
                   {'Terminals':len(solution.keys()),
                    'Title':'Construction Heuristic Result', 
                    'Conductor Number':t_nmbr, 
                    'Remained Nodes':len(G.nodes())-len(visited)
                    }
                   ]
    return return_list
```

```python
def meta_heu(G, sol_dic, max_speed = None):
    G_c = G.copy()
    path_list = []
    edge_list = []
    for k,v in sol_dic.items():
        for p in v:
            path_list.append(p)
            for e in [(p[n], p[n+1]) for n in range(len(p)-1)]:
                edge_list.append(e)
            
    for e in G.edges():
        if e not in edge_list:
            G_c.remove_edge(e[0],e[1])
    avg_speed = [-1, 0]
    ct = 0
    while avg_speed[-1] != avg_speed[-2]:
        ct += 1
        print ('number of meta-heuristic traverse time: ' + str(ct))

        new_total_dist = 0
        new_total_time = 0
        for e in G_c.edges():
            new_total_dist += G[e[0]][e[1]]['dist']
            new_total_time += G[e[0]][e[1]]['delt_t']
            
        new_avg_speed = new_total_dist/new_total_time
        avg_speed.append(new_avg_speed)
        for (n,m) in it.combinations(G_c.nodes(),2):
        
            if G_c.in_degree(n) == 1 and G_c.out_degree(n) == 1 and G_c.in_degree(m) == 1 and G_c.out_degree(m) == 1:
                (pre_n, suc_n) = (list(G_c.predecessors(n))[0], list(G_c.successors(n))[0])
                (pre_m, suc_m) = (list(G_c.predecessors(m))[0], list(G_c.successors(m))[0])
                dist_n_ori = G[pre_n][n]['dist'] + G[n][suc_n]['dist'] #这里用G而不是G_c
                dist_m_ori = G[pre_m][m]['dist'] + G[m][suc_m]['dist']
                [spd_pre_m_n,spd_n_suc_m,spd_pre_n_m,spd_m_suc_n] = [0,0,0,0]
                if (pre_m, n) in G.edges():
                    spd_pre_m_n = G[pre_m][n]['spd']
                if (n, suc_m) in G.edges():
                    spd_n_suc_m = G[n][suc_m]['spd']
                if (pre_n, m) in G.edges():
                    spd_pre_n_m = G[pre_n][m]['spd']
                if (m, suc_n) in G.edges():
                    spd_m_suc_n = G[m][suc_n]['spd']
                
                if spd_pre_m_n and spd_n_suc_m and spd_pre_n_m and spd_m_suc_n:
                    if max_speed:
                        if spd_pre_m_n <= max_speed and spd_n_suc_m <= max_speed and spd_pre_n_m <= max_speed and spd_m_suc_n <= max_speed:
                            if (pre_m, n) in G.edges() and (n, suc_m) in G.edges() and (pre_n, m) in G.edges() and(m, suc_n) in G.edges():
                                dist_n_prop = G[pre_m][n]['dist'] + G[n][suc_m]['dist']
                                dist_m_prop = G[pre_n][m]['dist'] + G[m][suc_n]['dist']
                                if dist_n_prop + dist_m_prop < dist_n_ori +dist_m_ori:
                                    print (str()+'meta-heuristic swapping executed between: ',(n,m))
                                    G_c.add_edge(pre_m, n, attr_dict = G.get_edge_data(pre_m, n))
                                    G_c.add_edge(n, suc_m, attr_dict = G.get_edge_data(n, suc_m))
                                    G_c.add_edge(pre_n, m, attr_dict = G.get_edge_data(pre_n, m))
                                    G_c.add_edge(m, suc_n, attr_dict = G.get_edge_data(m, suc_n))
                                    
                                    G_c.add_edge(pre_m, n, attr_dict = G.get_edge_data(pre_m, n))
                                    G_c.add_edge(n, suc_m, attr_dict = G.get_edge_data(n, suc_m))
                                    G_c.add_edge(pre_n, m, attr_dict = G.get_edge_data(pre_n, m))
                                    G_c.add_edge(m, suc_n, attr_dict = G.get_edge_data(m, suc_n))
                                    
    
                                    if suc_n == suc_m:
                                        G_c.remove_edge(pre_m,m)
                                        G_c.remove_edge(pre_n,n)
                                    else:
                                        if pre_n == pre_m:
                                            G_c.remove_edge(m,suc_m)
                                            G_c.remove_edge(n,suc_n)
                                        else:
                                            G_c.remove_edge(pre_m,m)
                                            G_c.remove_edge(pre_n,n)
                                            G_c.remove_edge(n,suc_n)
                                            G_c.remove_edge(m,suc_m)
                    if not max_speed:
                        if (pre_m, n) in G.edges() and (n, suc_m) in G.edges() and (pre_n, m) in G.edges() and(m, suc_n) in G.edges():
                            dist_n_prop = G[pre_m][n]['dist'] + G[n][suc_m]['dist']
                            dist_m_prop = G[pre_n][m]['dist'] + G[m][suc_n]['dist']
                            if dist_n_prop + dist_m_prop < dist_n_ori +dist_m_ori:
                                print (str()+'meta-heuristic swapping executed between: ',(n,m))
                                G_c.add_edge(pre_m, n, attr_dict = G.get_edge_data(pre_m, n))
                                G_c.add_edge(n, suc_m, attr_dict = G.get_edge_data(n, suc_m))
                                G_c.add_edge(pre_n, m, attr_dict = G.get_edge_data(pre_n, m))
                                G_c.add_edge(m, suc_n, attr_dict = G.get_edge_data(m, suc_n))
                                
                                G_c.add_edge(pre_m, n, attr_dict = G.get_edge_data(pre_m, n))
                                G_c.add_edge(n, suc_m, attr_dict = G.get_edge_data(n, suc_m))
                                G_c.add_edge(pre_n, m, attr_dict = G.get_edge_data(pre_n, m))
                                G_c.add_edge(m, suc_n, attr_dict = G.get_edge_data(m, suc_n))
                                

                                if suc_n == suc_m:
                                    G_c.remove_edge(pre_m,m)
                                    G_c.remove_edge(pre_n,n)
                                else:
                                    if pre_n == pre_m:
                                        G_c.remove_edge(m,suc_m)
                                        G_c.remove_edge(n,suc_n)
                                    else:
                                        G_c.remove_edge(pre_m,m)
                                        G_c.remove_edge(pre_n,n)
                                        G_c.remove_edge(n,suc_n)
                                        G_c.remove_edge(m,suc_m)

                        

    print (avg_speed[2:-1])
    solution = transfer_graph_to_solution_dic(G_c)
    visited = set()
    t_c_number = {}
    
    for t,p in solution.items():
        t_c_number[t] = len(p)
        for n in p:
            for i in n:
                visited.add(i)

    t_nmbr = 0
    for t in t_c_number:
        t_nmbr += t_c_number[t]
    
    return_list = [solution, 
                   {'terminals':len(solution.keys()),
                    'Title':'W.E. Swapping', 
                    'conductor number':t_nmbr, 
                    'remained W.E.':len(G.nodes())-len(visited)
                    }
                   ]
    return return_list
```

```python
def avg_speed(G, sol_dic):
    G_c = G.copy()
    routes_in_edge = []
    solution_l = []
    for k,v in sol_dic.items():
        for r in v:
            solution_l.append(r)
    
    for p in solution_l:
        routes_in_edge.append([(p[n],p[n+1]) for n in range(len(p)-1)])   

    ttl_dist = 0
    ttl_time = 0
    for p in routes_in_edge:
        for e in p:
            ttl_dist += G_c[e[0]][e[1]]['dist']
            ttl_time += G_c[e[0]][e[1]]['delt_t']
    
    avg_spd = ttl_dist/ttl_time
    return avg_spd
```

```python
def plot_t_x_plot(G_origin, solu, dot_line=False, plt_ttl = True, width = 1200, height = 900):
    attr_dic = solu[1]
    routes_in_edge = []
    solution_l = []
    for k,v in solu[0].items():
        for r in v:
            solution_l.append(r)
    for p in solution_l:
        routes_in_edge.append([(p[n],p[n+1]) for n in range(len(p)-1)])
    
    colors =        ['r',   'b',    'y',    'g',    'c',    'm',    'k']
    linewidths =    [5 for n in routes_in_edge]
    
    
    node_label_dict = {}
    pos_dict = {}
    
    for (n,attr) in G_origin.nodes(data=True):
        (node_label_dict[n], pos_dict[n]) = (attr['label'], attr['coord'])
    
    
    G_draw_edge_label = G_origin.copy()
    if dot_line:
    #-------plot the dashlines----#
        nx.draw_networkx(G_origin,
                         labels =node_label_dict,
                         edgelist = G.edges(), 
                         pos = pos_dict, 
                         edge_color = 'purple', 
                         style = 'dashed', 
                         node_color = 'pink',
                         arrow = False
                         )
    #-------plot the dashlines----#
    
    #-------plot the conductor trips----#
    for ct, route in enumerate(routes_in_edge):
       
        if route[0][0] == 'T0Begin':
            color = 'y'
        if route[0][0] == 'T1Begin':
            color = 'b'
        if route[0][0] == 'T2Begin':
            color = 'r'
        if route[0][0] == 'T3Begin':
            color = 'k'
        
        nx.draw_networkx(G, 
                         labels = node_label_dict, 
                         edgelist = route, 
                         pos = pos_dict, 
                         width = linewidths[ct],
                         edge_color = color,
                         alpha = 0.5, 
                         arrows = False)
    #-------plot the conductor trips----#
    #-------plot edge speed-------------#
    edges_to_draw = []
    
    for r in routes_in_edge:
        for e in r:
            edges_to_draw.append(e)
    
    for e in G_origin.edges():
        if e not in edges_to_draw:
            G_draw_edge_label.remove_edge(e[0],e[1])
    plt.get_current_fig_manager().resize(width, height)
    spd_label = nx.get_edge_attributes(G_draw_edge_label, 'spd')
    
    nx.draw_networkx_edge_labels(G_draw_edge_label,  
                                 pos=pos_dict,
                                 edge_labels = spd_label, 
                                 alpha = 15.0)
    
      
    if plt_ttl:
        ttl = ''
        for k,v in attr_dic.items():
            if k == 'Title':
                ttl = ttl + k +':  ' + str(v) +';'
            if k == 'terminals':
                plt.text(terminal_loc[-1]/2, shift_hour+0.5, k+':  '+str(v))
            if k == 'conductor number':
                plt.text(terminal_loc[-1]/2, shift_hour+1, k+':  '+str(v))
            if k == 'remained W.E.':
                plt.text(terminal_loc[-1]/2, shift_hour+1.5, k+':  '+str(v))
            
        plt.title(ttl)
    
            
    
    plt.show()
    
    #-------plot edge speed-------------#
```
